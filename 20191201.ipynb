```python
passwordbook=[]
def print_menu():
    print("="*30)
    print("密码薄")
    print("1.添加网址信息")
    print("2.删除网址信息")
    print("3.修改网址信息")
    print("4.显示所有网址信息")
    print("5.查询网址信息")
    print("6.保存数据")
    print("0.退出系统")
    print("="*30)

#增加一个新的网址和密码
def add_info():
    new_web = input("请输入要存储的网址：")
    new_password = input("请输入该网址对应的密码：")
    new_infos = {}
    new_infos['web'] = new_web
    new_infos['password'] = new_password
    passwordbook.append(new_infos)

#删除网站
def del_info(web):
    del_name = int(input("请输入要删除的网站的序号：")) - 1
    del web[del_name]
#修改网站信息
def modify_info():
    web_id=int(input("请输入要修改的网址的序号："))
    web=input("请输入修改后的网址的名称：")
    password=input("请输入新网址的密码：")
    passwordbook[web_id -1]['web'] = web
    passwordbook[web_id -1]['password'] = password

#显示网站信息
def show_infos():
    print("=" * 30)
    print("网址的信息如下：")
    print("=" * 30)
    print("序号  网址名称  网址密码  ")
    i = 1
    for temp in passwordbook:
        print("%d  %s  %s  " % (i, temp['web'], temp['password']))
        i += 1
#查询网址信息
def select_infos():
    select_web=input("请输入您要查询的网址名称：")
    print("您要查询的网址信息如下：")
    length=len(passwordbook)
    for i in (0,length-1):
            if(select_web in passwordbook[i].values()):
                print(i+1,passwordbook[i]['web'],passwordbook[i]['password'])
#保存所有的学生信息到文件中
def save_to_file():
    file=open("password.txt","w+")
    file.write(str(passwordbook))
    file.close()
#恢复数据    z
def recover_data():
    global students_infos
    file=open("password.txt")
    content=file.read()
    students_infos=eval(content)
    file.close()

def main():
    recover_data()
    while True:
        print_menu()  # 打印菜单
        key = input("请输入功能对应的数字：")  # 获得用户输入的序号
        if key == '1':  # 添加学生的信息
            add_info()
        elif key == '2':
            del_info(passwordbook)
        elif key == '3':
            modify_info()
        elif key == '4':
            show_infos()
        elif key == '5':
            select_infos()
        elif key == '6':
            save_to_file()
```
